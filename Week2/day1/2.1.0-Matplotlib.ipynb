### A Short Visit to Matplotlib and Displaying Images in Jupyter Notebooks

Matplotlib is a powerful plotting library in Python that allows you to create a wide variety of static, animated, and interactive plots. It's particularly useful for visualizing data and displaying images.

#### Installing Matplotlib

To install Matplotlib, you can use pip:

#### Importing Necessary Libraries

First, let's import the libraries we'll be using:


````python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
````

The `%matplotlib inline` magic command is used to display plots inline within the Jupyter Notebook.


#### Reading and Displaying Images with Matplotlib

This is how you can read and display an image using Matplotlib:


````python
import cv2
import matplotlib.pyplot as plt

# Read an image using OpenCV
img = cv2.imread('images/nascom.jpeg')

# Convert the image from BGR to RGB (OpenCV reads images in BGR format)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Display the image using Matplotlib
plt.imshow(img_rgb)
plt.title('TECHNOLOGY FOR ALL')
plt.axis('off')  # Hide axes
plt.show()
````

#### Displaying Multiple Images

You can display multiple images in a single plot using subplots:


````python
import cv2
import matplotlib.pyplot as plt

# Read multiple images
img1 = cv2.imread('images/nascom.jpeg')
img2 = cv2.imread('images/nascom2.jpeg')

# Convert images from BGR to RGB
img1_rgb = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2_rgb = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

# Create subplots
fig, axes = plt.subplots(2,2, figsize=(10, 15))

# Display the first image
axes[0][0].imshow(img1_rgb)
axes[0][0].set_title('Component')
axes[0][0].axis('off')

axes[0][1].axis("off")

# Display the second image
axes[1][0].imshow(img2_rgb)
axes[1][0].set_title('Website')
axes[1][0].axis('off')

axes[1][1].set_title("Empty")

# Show the plot
plt.show()
````

#### Adding Color Maps

Matplotlib provides various color maps that you can use to display grayscale images:


````python
import cv2
import matplotlib.pyplot as plt

# Read an image in grayscale
img_gray = cv2.imread('images/nascom2.jpeg', cv2.IMREAD_GRAYSCALE)
# print(img_gray.shape)

# Display the grayscale image with a color map
plt.imshow(img_gray, cmap='Greens')
plt.title('Hot Image')
plt.axis('off')
plt.show()
````
